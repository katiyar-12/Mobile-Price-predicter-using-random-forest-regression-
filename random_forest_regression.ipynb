# Random Forest Regression

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('mobile_dataset.csv' , encoding='ISO-8859-1')
dataset.head()
```

# ** preparing the dataset ** #

```python
ram = dataset['RAM'].str.extract(r'(\d+\.?\d*)').astype(float)

front_camera = dataset['Front Camera'].str.extract(r'(\d+\.?\d*)').astype(float)

back_camera = dataset['Back Camera'].str.extract(r'(\d+\.?\d*)').astype(float)

battery = dataset['Battery Capacity'].str.extract(r'(\d+\.?\d*)').astype(float)

screen_size = dataset['Screen Size'].str.extract(r'(\d+\.?\d*)').astype(float)




```

```python
X = pd.concat([ram , front_camera , back_camera , battery , screen_size ] , axis=1)
X.columns = ['RAM' , 'Front Camera' , 'Back Camera' , 'Battery Capacity' ,'Screen Size' ]
X = X.values

y = dataset.iloc[: , 10].str.replace("INR" ,"" , regex=False).str.replace("," , "" , regex=False ).str.strip().astype(float).values
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Training the Random Forest Regression model on the whole dataset

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(X_train, y_train)
```

## Predicting the Test set results

```python
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

## Evaluating the Model Performance

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
